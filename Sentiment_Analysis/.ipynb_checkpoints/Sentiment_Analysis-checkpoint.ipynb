# Sentiment Analysis

```python
import pandas as pd
import regex as re
import spacy
import unidecode
import xlsxwriter

from pysentimiento import create_analyzer

path_to_MyModule = '..'

import sys
sys.path.insert(0, path_to_MyModule) 

from MyModule.NLPFunctions import *
```

```python
df = pd.read_excel('datos.xlsx')[['ID','texto','desafio']]
```

```python
# Limpieza
df.drop_duplicates(subset='texto', inplace=True)

# Quitando texto de mas en columna "desafio"
df['desafio'] = df['desafio'].apply(lambda x: re.findall('[0-9]+', x)[0])

# A str
df['texto'] = df['texto'].astype(str)
```

## Pre process and tokenization

```python
%%capture
!python -m spacy download es_core_news_md
```

```python
pp = Preprocess(lemma=False, stopwords=False)
df['prepro_text'] = pp.preprocess(df['texto'])
df['prepro_text'] = df['prepro_text'].apply(lambda x: ' '.join(x))
```

# Iteratation

```python
def analyse_sentiment_3d(phrase, sent = ['POS', 'NEG', 'NEU']):
    
    """ Predicts sentiment of phrase in three dimensions: POS, NEG, NEU. """
    
    analyzer = create_analyzer(task="sentiment", lang="es")
    
    predictions = {}
    for s in sent:
        prediction = analyzer.predict(phrase).probas[s]
        predictions[s] = prediction
    
    return predictions
```

```python
emotions_dict = {}

for phrase in df['prepro_text'].values:
    new_dict = analyse_sentiment_3d(phrase)
    
    for key, value in new_dict.items():
        if key in emotions_dict:
            emotions_dict[key].append(value)
        else:
            emotions_dict[key] = [value]
```

```python
df['POS'] = emotions_dict['POS']
df['NEG'] = emotions_dict['NEG']
df['NEU'] = emotions_dict['NEU']
```

```python
df.to_excel('sentiment_analysis.xlsx', 
            index=False, 
            header=True, 
            startrow=1, 
            startcol=1, 
            freeze_panes=(2,2), 
            engine='xlsxwriter')
```
